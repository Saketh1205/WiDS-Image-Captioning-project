```python
#converting to lower case
test_data['Sentence'] = test_data['Sentence'].str.lower()
test_data
```

```python
#Tokenization
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
```

```python
def tokenize(column):
    tokens = nltk.word_tokenize(column)
    return [w for w in tokens if w.isalpha()]
```

```python
test_data['tokenized'] = test_data.apply(lambda x: tokenize(x['Sentence']), axis=1)
test_data[['tokenized']]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#stop_words = stopwords.words('english')
#test_data['tokenized'] = test_data['tokenized'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
def remove_english_stopwords_func(text):
    # check in lowercase 
    t = [token for token in text if token.lower() not in stopwords.words("english")]
    text = ' '.join(t)    
    return text
```

```python
test_data['Stop_Words'] = test_data['tokenized'].apply(remove_english_stopwords_func)
```

```python
test_data
```

```python
#Lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word,pos='v') for word in words]
    return ' '.join(words)
test_data['lemmatized'] = test_data['Stop_Words'].apply(lemmatize_words)
```

```python
test_data
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
test_data['sentiment_scores'] = None
```

```python
for i in range(len(test_data)):
    sentiment = sia.polarity_scores(test_data.iloc[i]['lemmatized'])
    test_data.at[i, 'sentiment_scores'] = sentiment
```

```python
test_data
```

```python
for i in range(len(test_data)):
    sentiment = sia.polarity_scores(test_data.iloc[i]['lemmatized'])
    if sentiment['compound'] >= 0.43:
        test_data.at[i, 'pred_sentiment'] = 'positive'
    elif sentiment['compound'] <= -0.43:
        test_data.at[i, 'pred_sentiment'] = 'negative'
    else:
        test_data.at[i, 'pred_sentiment'] = 'neutral'
```
